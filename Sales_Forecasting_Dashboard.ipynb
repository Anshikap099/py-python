**8. MACHINE LEARNING MODEL**

```
url = "https://docs.google.com/spreadsheets/d/1NI1A7GZfFPKfQmLyYm_xknCTuKGF0fz_/export?format=csv"
df = pd.read_csv(url)

df.columns = df.columns.str.lower()
print("DataFrame reloaded. First 5 rows:")
display(df.head())
```

```
df = df.drop(columns=['ind1', 'ind2'], errors='ignore')
df['returns'] = df['returns'].fillna(0)

df['date'] = pd.to_datetime(df['order date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
print("Data cleaning and feature engineering re-applied.")
print(f"DataFrame shape after cleaning: {df.shape}")
```

```
region_selected = "All"  # change to Central/East/West if needed

if region_selected != "All":
    df = df[df['region'] == region_selected]
print(f"Region filter re-applied. Current region selection: {region_selected}")
print(f"DataFrame shape after filtering: {df.shape}")
```

```
monthly = df.groupby(['year','month'])['sales'].sum().reset_index()
category = df.groupby('category')['sales'].sum().reset_index()
region = df.groupby('region')['sales'].sum().reset_index()
print("Aggregated DataFrames (monthly, category, region) re-prepared.")
```

```
from sklearn.ensemble import RandomForestRegressor

X = df[['year','month']]
y = df['sales']

model = RandomForestRegressor()
model.fit(X,y)
```

```
future_years = [2021] * 12
future_months = range(1, 13)
forecast_input = pd.DataFrame({
    'year': future_years,
    'month': future_months
})
print("Forecast input data created for the next 12 months.")
display(forecast_input.head())
```

```
forecast = model.predict(forecast_input)
print("Sales forecast generated using the trained model.")
```

**9. FORECAST**

```
forecast_df = pd.DataFrame({
    'Month': range(1, 13),
    'Predicted Sales': forecast
})
print("Forecast DataFrame created.")
display(forecast_df.head())
```

```
import pandas as pd

frames = []

for i in range(1, 13):
    temp = forecast_df.iloc[:i].copy()
    temp['frame'] = i
    frames.append(temp)

animated_df = pd.concat(frames)
```

```
import plotly.express as px

fig = px.line(
    animated_df,
    x='Month',
    y='Predicted Sales',
    animation_frame='frame',
    markers=True,
    title="🔮 Animated Sales Forecast"
)

fig.update_layout(
    template="plotly_dark",
    transition={'duration': 500}
)

fig.show()
```

**10. FINAL DASHBOARD**

```
dashboard = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Monthly", "Category", "Region", "Forecast"),
    specs=[[{"type":"scatter"},{"type":"bar"}],
           [{"type":"pie"},{"type":"scatter"}]]
)

dashboard.add_trace(go.Scatter(x=monthly['month'], y=monthly['sales']),1,1)
dashboard.add_trace(go.Bar(x=category['category'], y=category['sales']),1,2)
dashboard.add_trace(go.Pie(labels=region['region'], values=region['sales']),2,1)
dashboard.add_trace(go.Scatter(x=forecast_df['Month'], y=forecast_df['Predicted Sales']),2,2)

dashboard.update_layout(
    height=700,
    template="plotly_dark",
    title="🔥 Final Sales Dashboard"
)

dashboard.show()
```
